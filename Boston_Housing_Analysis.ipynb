```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv")
```

```python
# Гистограмма для медианной стоимости домов (MEDV)
plt.figure(figsize=(8, 5))
sns.histplot(data["MEDV"], bins=30, kde=True, color="blue")
plt.title("Гистограмма медианной стоимости домов")
plt.xlabel("Стоимость (тыс. долларов)")
plt.ylabel("Частота")
plt.show()
```

```python
# Гистограмма для реки Charles (CHAS)
plt.figure(figsize=(6, 4))
sns.countplot(x=data["CHAS"], hue=data["CHAS"], palette="Set2", legend=False)
plt.title("Гистограмма реки Charles")
plt.xlabel("Присутствие (1) / Отсутствие (0)")
plt.ylabel("Частота")
plt.show()
```

```python
# Гистограмма MEDV в зависимости от возраста (AGE)
age_bins = [0, 35, 70, 100]
age_labels = ["<=35", "35-70", ">70"]
data["AGE_group"] = pd.cut(data["AGE"], bins=age_bins, labels=age_labels)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data, x="MEDV", hue="AGE_group", multiple="stack", bins=30)
plt.title("Гистограмма стоимости домов в зависимости от возраста")
plt.xlabel("Стоимость (тыс. долларов)")
plt.ylabel("Частота")
plt.show()
```

```python
# Диаграмма рассеяния NOX vs. INDUS
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data["INDUS"], y=data["NOX"], alpha=0.7)
plt.title("Зависимость концентрации NOX от процента незастроенной земли")
plt.xlabel("Процент незастроенной земли")
plt.ylabel("Концентрация NOX")
plt.show()
```

```python
# Гистограмма для соотношения учеников и учителей (PTRATIO)
plt.figure(figsize=(8, 5))
sns.histplot(data["PTRATIO"], bins=20, kde=True, color="purple")
plt.title("Гистограмма соотношения учеников и учителей")
plt.xlabel("Соотношение учеников к учителям")
plt.ylabel("Частота")
plt.show()
```
